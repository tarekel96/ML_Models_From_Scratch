```python
# import LinearRegression model
from linear_regression import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
# produce linear data
x = np.random.randint(100, size=100)/30 - 2
X = x.reshape(-1, 1)
# y = x + 2
y = x + 2 + 0.3*np.random.randn(100)
```

```python
x.shape, y.shape
```

```python
x = x.reshape(-1, 1)
```

```python
x.shape
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y)
ax.set_title('Simple Linear Data')
```

```python
lr_model = LinearRegression(epochs=100)
```

```python
lr_model.fit(x, y)
```

```python
lr_model.w
```

```python
y_hat = lr_model.predict(x)
```

```python
costs = lr_model.costs
```

```python
fig, ax = plt.subplots()
ax.scatter(np.arange(1,101,1), costs)
ax.set_title('Gradient Descent')
ax.set_xlabel('epoch')
ax.set_ylabel('loss')
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y)
ax.plot(x,y_hat, c='red')
ax.set_title('LR Model predictions')
```
